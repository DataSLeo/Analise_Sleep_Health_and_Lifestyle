## 3. Análise exploratória de dados

### 3.1. Qual é a média das idades dos gêneros?

```python
import matplotlib.pyplot as plt
```

```python
# Data Mean Age
dma = data.groupby('Gender')['Age'].mean().astype(int)
dma
```

```python
plt.bar(dma.index, dma.values, color='mediumspringgreen')

for i in range(len(dma)):
  x_ = dma.index[i]
  y_ = dma.values[i]
  plt.text(x_, y_, f'{y_}')

plt.xlabel("Gênero")
plt.ylabel("Idade")
plt.title("Média de idades por gênero")
plt.show()
```

Aqui mostra que a média de idades mais predominantes são os femininos.

### 3.2. Qual gênero predomina mais?

```python
# Data Gender
dg = data.groupby('Gender')['Gender'].size()
dg
```

```python
plt.bar(dg.index, dg.values, color='mediumspringgreen')

for i in range(len(dg)):
  x_ = dg.index[i]
  y_ = dg.values[i]
  plt.text(x_, y_, f'{y_}')

plt.xlabel("Gêneros")
plt.ylabel("Quantidade")
plt.title("Quantidade de indivíduos por gênero")
plt.show()
```

O gênero mais predominante é o masculino.

### 3.3. Qual gênero é mais estressado?

```python
# Data Gender Stress
dgsm = data[data['Gender'] == 'Male']
dgsm = dgsm[['Stress Level']]
dgsm = dgsm.groupby('Stress Level')['Stress Level'].size()
dgsm
```

```python
plt.bar(dgsm.index, dgsm.values, color='mediumspringgreen')

for i in range(len(dgsm)):
  x_ = dgsm.index[i]
  y_ = dgsm.values[i]
  plt.text(x_, y_, f'{y_}', ha='center')

plt.xlabel("Níveis")
plt.ylabel("Quantidade")
plt.title("Quantidade de níveis por gênero masculino")
plt.show()
```

```python
dgsm_size = 0

for x in dgsm:
  dgsm_size += x

dgsm_percent = []

for x in dgsm:
  dgsm_percent.append(x/dgsm_size * 100)
```

```python
plt.bar(dgsm.index, dgsm.values, color='mediumspringgreen')

for i in range(len(dgsm)):
  x_ = dgsm.index[i]
  y_ = dgsm_percent[i]
  plt.text(x_, y_, f'{int(y_)}%', ha='center', va='top')

plt.xlabel("Níveis")
plt.ylabel("Quantidade")
plt.title("Quantidade de níveis por gênero masculino")
plt.show()
```

Em relação ao gênero masculino mostra que 34% do estresse está no nível 5.

```python
# Data Gender Stress
dgsf = data[data['Gender'] == 'Female']
dgsf = dgsf[['Stress Level']]
dgsf = dgsf.groupby('Stress Level')['Stress Level'].size()
dgsf
```

```python
plt.bar(dgsf.index, dgsf.values, color='mediumspringgreen')

for i in range(len(dgsm)):
  x_ = dgsf.index[i]
  y_ = dgsf.values[i]
  plt.text(x_, y_, f'{y_}', ha='center')

plt.xlabel("Níveis")
plt.ylabel("Quantidade")
plt.title("Quantidade de níveis por gênero feminino")
plt.show()
```

```python
dgsf_size = 0

for x in dgsf:
  dgsf_size += x

dgsf_percent = []

for x in dgsf:
  dgsf_percent.append(x/dgsf_size * 100)
```

```python
plt.bar(dgsf.index, dgsf.values, color='mediumspringgreen')

for i in range(len(dgsf)):
  x_ = dgsf.index[i]
  y_ = dgsf_percent[i]
  plt.text(x_, y_, f'{int(y_)}%', ha='center', va='center')

plt.xlabel("Níveis")
plt.ylabel("Quantidade")
plt.title("Quantidade de níveis por gênero feminino")
plt.show()
```

Em relação ao gênero feminino mostra que 36% do estresse está no nível 3.

Entre esses gráficos mostra que o gênero masculino tende a ser mais estressado do que o gênero feminino.

### 3.4. Qual profissão predomina mais?

```python
# Data Occupation
do = data.groupby('Occupation')['Occupation'].size()
do = do.sort_values(ascending=False)
do
```

```python
plt.bar(do.index, do.values, color='mediumspringgreen')

for i in range(len(do)):
  x_ = do.index[i]
  y_ = do.values[i]
  plt.text(x_, y_, f'{y_}', ha='center')

plt.xlabel("Profissão")
plt.ylabel("Quantidade")
plt.title("Quantidade de profissão total")
plt.xticks(rotation=90)
plt.show()
```

Aqui nos mostra que a profissão nurse (enfermeira) possui a maior frequência.
